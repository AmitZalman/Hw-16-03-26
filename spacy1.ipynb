```python
import spacy
nlp = spacy.load("en_core_web_sm")

# The model response is a SpaCy Doc object
doc = nlp("Tesla is looking at buying U.S. startup for $6 million")

for token in doc:
    print(f'[{token.pos:3}]', token.text, token.pos_, token.dep_)
```

```python
#1. Named Entities Extraction

import spacy
nlp = spacy.load("en_core_web_sm")

# The model response is a SpaCy Doc object
doc = nlp("Taylor Swift performed in Los Angeles on March 3rd, 2023")

for entity in doc.ents:
    print(entity.text, entity.label_)
```

```python
# 2. Entity Classification

import spacy
nlp = spacy.load("en_core_web_sm")


doc = nlp("Serena Williams had dinner with Tom Hanks in Paris.")

for entity in doc.ents:
    if entity.label_ == 'PERSON':
        print(entity.text, entity.label_)
    
```

```python
#3. Lemmatization

import spacy
nlp = spacy.load("en_core_web_sm")


doc = nlp("She was running and had run 5 kilometers by 7am")

for token in doc:
    print(f"{token.text:<15} -> {token.lemma_}")
```

```python
#4. Stop Word Removal

import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp("This is an example sentence with some stop words")

for token in doc:
    if not token.is_stop:
        print(f"{token.text}")
```

```python
#5. Custom Stop Word

import spacy
nlp = spacy.load("en_core_web_sm")
nlp.vocab["powerful"].is_stop = True

doc = nlp("SpaCy is awesome and powerful")

for token in doc:
    if token.is_stop:
        print(f"{token.text}")
```

```python
#6. Phrase Matcher

import spacy
from spacy.matcher import PhraseMatcher

nlp = spacy.load("en_core_web_sm")
matcher = PhraseMatcher(nlp.vocab)

patterns = [nlp("Artificial Intelligence"), nlp("artificial intelligence")]

matcher.add("AI_PATTERN", patterns)

doc = nlp("Artificial Intelligence is the future. I study artificial intelligence")
matches = matcher(doc)

for match_id, start, end in matches:
    print(doc[start:end].text)
```

```python
#7. POS Tagging + Explanation


import spacy
nlp = spacy.load("en_core_web_sm")


doc = nlp("The cat sat on the mat")

for token in doc:
    pos_explanation = spacy.explain(token.pos_)
    
    print(f"{token.text:<5} {token.pos_:<5} {pos_explanation}")
```

```python
#8. POS Tagging + Displacy Visualization

from spacy import displacy
import spacy
nlp = spacy.load("en_core_web_sm")


doc = nlp("Apple is looking at buying a U.K. startup for $1 billion")

for token in doc:
    pos_explanation = spacy.explain(token.pos_)
    
    print(f"{token.text:<5} {token.pos_:<5} {pos_explanation}")
displacy.render(doc, style="dep", jupyter=True)
```
